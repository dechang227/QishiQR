```python
import matplotlib.pyplot as plt
```

```python
symbol = 'bu'
instrument = 'bu1606'
topdir = 'D:\GitHub\QishiQR\Data'
offset = [0.1]
freq = 5
```

```python
tick_day = df_reader(instrument + '*', topdir=topdir + '/'+symbol+ '/day', offset=offset, freq=str(freq)+'min',day=True, symbol=symbol).get_tick(raw=False)
tick_night = df_reader(instrument + '*', topdir=topdir + '/'+symbol+ '/night', offset=offset, freq=str(freq)+'min',day=False, symbol=symbol).get_tick(raw=False)
```

```python
tick_all = pd.concat([tick_day['0.1'], tick_night['0.1']])
tick_all.sort_index(inplace=True)
```

```python
sub_tick = tick_all#[tick_all.index <= pd.to_datetime('2016-2-1')]

fig = plt.figure()
plt.plot(sub_tick.index, sub_tick.LastPrice, label="LastPrice")
plt.plot(sub_tick.index, sub_tick.AskPrice1, label="AskPrice")
plt.plot(sub_tick.index, sub_tick.BidPrice1, label="BidPrice")
plt.legend(loc='upper right')
plt.title('Equity Curve')
plt.show()
```

```python
tick_all['is_in'] = tick_all.apply(lambda row: 1 if (row['LastPrice'] <= row['AskPrice1'] and row['LastPrice'] >= row['BidPrice1']) else 0, axis=1) 
```

```python
fig = plt.figure()
plt.plot(tick_all.index, tick_all.is_in)
plt.show()
```

```python
tick_all['spread'] = tick_all['AskPrice1'] - tick_all['BidPrice1']
```

```python
fig = plt.figure()
plt.plot(tick_all.index, tick_all.spread)
plt.show()
```
