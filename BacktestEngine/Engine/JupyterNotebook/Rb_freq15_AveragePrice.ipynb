```python
Rb_Results = GetResult("../Results/Rb_freq15_AveragePrice/")
```

## Rb - Group test results

```python
import pandas as pd
Result_directory = pd.DataFrame(columns=["price_threshold","fixed_cost","frequency","offset","tester"])

for result in Rb_Results:
    tmp_entry = {
        "price_threshold": result['localPara']['price_threshold'],
        "fixed_cost": result["localPara"]["fixed_cost"],
        "frequency": result["localPara"]["frequency"],
        "offset": result["localPara"]["offset"],
        "tester": result
    }
    Result_directory = Result_directory.append(tmp_entry,ignore_index=True)
    
```

```python
Parameters = Result_directory[["price_threshold","fixed_cost","frequency"]].drop_duplicates()
```

```python
for _, row in Parameters.iterrows():
    Offset_Average_results = OffSetAverage(
        Result_directory.query("price_threshold=={} and fixed_cost=={}  \
        and frequency=={}".format(row['price_threshold'], row['fixed_cost'], row['frequency'])).tester.values)
    print("="*72)
    print("price_threshold=={} and fixed_cost=={} and frequency=={}"
          .format(row['price_threshold'], row['fixed_cost'], row['frequency']))
    Benchmark = Offset_Average_results.GetBenchmark()
    AveEqCurve = Offset_Average_results.GetAveEquityCurve()

    # Plot
    fig, ax = plt.subplots(1,2, figsize=(20,6))
    Benchmark['insample'].plot(ax=ax[0],label='b')
    AveEqCurve['insample'].plot(ax=ax[0])
    ax[0].legend(['b',2,3,4,5,6,7,8])
    ax[0].set_title('Equity_Curve - insample - OffsetAverage')

    Benchmark['outsample'].plot(ax=ax[1],label='b')
    AveEqCurve['outsample'].plot(ax=ax[1])
    ax[1].legend(['b',2,3,4,5,6,7,8])
    ax[1].set_title('Equity_Curve - Outsample - OffsetAverage')
    plt.show()
    
    display(Offset_Average_results.GetAvePerformance()['insample'])
    display(Offset_Average_results.GetAvePerformance()['outsample'])
```
