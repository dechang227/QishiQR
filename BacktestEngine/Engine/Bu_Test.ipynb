```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import sys
sys.path.append("/home/runmin/Documents/Qishi/QishiQR/BacktestEngine")

from glob import glob
import pickle
import matplotlib.pyplot as plt
import Backtesting



def GetResult(RESULT_DIR):
    files = sorted(glob(RESULT_DIR+"/*.pkl"))

    Tester_Results = []
    for idx, file in enumerate(files):
        with  open(file,"rb") as f:
            Tester_Results.append(pickle.load(f))

        print("\n================================================================")
        params = Tester_Results[-1]["globalPara"]
        print("file_path = ", file)

        for key, item in Tester_Results[-1]["localPara"].items():
            print("{} = {}".format(key, item))

        fig, ax = plt.subplots(1,2, figsize=(16,8))
        Tester_Results[-1]["insample"].ensemble.ensembles[0].result.benchmark.plot(ax=ax[0])
        Tester_Results[-1]["insample"].plot(ax=ax[0])
        Tester_Results[-1]["outsample"].ensemble.ensembles[0].result.benchmark.plot(ax=ax[1])
        Tester_Results[-1]["outsample"].plot(ax=ax[1])

        ax[0].legend().set_visible(False)
        ax[0].set_xlabel("Date")
        ax[0].set_title("Equity curve - Insample")
        ax[1].set_xlabel("Datae")
        ax[1].set_title("Equity curve - Outsample")
        plt.show()


    
```
