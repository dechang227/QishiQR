```python
import sys
sys.path.append("/home/runmin/Documents/Qishi/QishiQR/BacktestEngine")

from glob import glob
import pickle
import matplotlib.pyplot as plt
import Backtesting
import pandas as pd
```

#  10min Frequeny - Train: AvePrice2,  Signal: AvePrice2


```python
result_folder = './Results_train_ave_signal_ave/'
files = glob(result_folder+"*.pkl")



Tester_Results = []

for idx, file in enumerate(files):
    with  open(file,"rb") as f:
        Tester_Results.append(pickle.load(f))

    print("\n=======================")
    params = Tester_Results[-1]["globalPara"]
    print(idx,":{syb}".format(syb=params.symbol))
    print(Tester_Results[-1]["localPara"])
    fig, ax = plt.subplots()
    Tester_Results[-1]["model"].ensemble.ensembles[0].result.benchmark.plot(ax=ax)
    Tester_Results[-1]["model"].plot(ax=ax)
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
    plt.xlabel("Date")
    plt.show()
```

#### Offset average

```python
Time_index = Tester_Results[0]["model"].ensemble.ensembles[0].result.index
Averaged_return = pd.DataFrame(index=Time_index)
Averaged_return["Benchmark"] = Tester_Results[0]["model"].ensemble.ensembles[0].result.benchmark

for idx, model_order in enumerate([2,3,4,5,6,7,8]):
    equity_curve = 0
    for offset_idx, tester in enumerate(Tester_Results, start=1):
        equity_curve += tester["model"].ensemble.ensembles[idx].result.equitycurve.values
    Averaged_return[model_order] = equity_curve/offset_idx
```

```python
fig = plt.figure(figsize=(10,6))
ax = plt.gca()
Averaged_return.plot(ax=ax)
ax.set_title("Freq=5min, Offset averaged")
ax.set_ylabel("Equity Curve")
ax.set_xlabel("Date")
```
